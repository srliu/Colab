#Pandas 教程 10(Datetime-Timeseries)

## 一些常用的日期時間字符串格式碼：
* %Y：四位數的年份，例如 2022。
* %m：兩位數的月份，例如 01。
* %d：兩位數的日期，例如 05。
* %H：兩位數的小時（24 小時制），例如 09。
* %I：兩位數的小時（12 小時制），例如 09。
* %M：兩位數的分鐘，例如 30。
* %S：兩位數的秒，例如 59。
* %f：微秒（六位數），例如 000000。
* %z：時區偏移量，例如 +0800。
* %Z：時區名稱，例如 CST。
* %j：一年中的第幾天，例如 365。
* %U：一年中的第幾周（以星期日為一周的第一天），例如 52。
* %W：一年中的第幾周（以星期一為一周的第一天），例如 52。
* %c：完整的日期時間，例如 Mon Sep 30 07:06:05 2013。
* %x：日期部分，例如 09/30/13。
* %X：時間部分，例如 07:06:05。
